# Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Import the datasets

```python
evo = pd.read_csv('https://raw.githubusercontent.com/sohanjs111/robots-speed-model/main/Dataset/Evocortex.csv')
bot = pd.read_csv('https://raw.githubusercontent.com/sohanjs111/robots-speed-model/main/Dataset/Turtlebot3.csv')
```

## For TurtleBot3 

```python
bot_x = bot.iloc[:, 0].values    # Speed
bot_y1 = bot.iloc[:, 1].values   # Green Carpet
bot_y2 = bot.iloc[:, 2].values   # Floor
```

```python
bot_x = np.reshape(bot_x, (-1, 1))
```

### Plots

```python
plt.scatter(bot_x, bot_y1, color = 'red')
plt.title('Turtlebot3 for Green carpet')
plt.xlabel('Speed in m/s')
plt.ylabel('Standard Deviation')
plt.show()
```

```python
plt.scatter(bot_x, bot_y2, color = 'red')
plt.title('Turtlebot3 for Floor')
plt.xlabel('Speed in m/s')
plt.ylabel('Standard Deviation')
plt.show()
```

## Evocortex

```python
evo_x = evo.iloc[:, 0].values    # Speed
evo_y1 = evo.iloc[:, 1].values   # Green Carpet
evo_y2 = evo.iloc[:, 2].values   # Floor
evo_y3 = evo.iloc[:, 3].values   # Foam
```

```python
evo_x = np.reshape(evo_x, (-1, 1))
```

### Plots

```python
plt.scatter(evo_x, evo_y1, color = 'blue')
plt.title('Evocortex for Green carpet')
plt.xlabel('Speed in m/s')
plt.ylabel('Standard Deviation')
plt.show()
```

```python
plt.scatter(evo_x, evo_y2, color = 'blue')
plt.title('Evocortex for floor')
plt.xlabel('Speed in m/s')
plt.ylabel('Standard Deviation')
plt.show()
```

```python
plt.scatter(evo_x, evo_y3, color = 'blue')
plt.title('Evocortex for foam')
plt.xlabel('Speed in m/s')
plt.ylabel('Standard Deviation')
plt.show()
```

# TurtleBot3

## Simple Linear Regression

### models 

```python
from sklearn.linear_model import LinearRegression
bot_sr1 = LinearRegression()
bot_sr1.fit(bot_x, bot_y1)
```

```python
from sklearn.linear_model import LinearRegression
bot_sr2 = LinearRegression()
bot_sr2.fit(bot_x, bot_y2)
```

### Plots

```python
plt.scatter(bot_x, bot_y1, color = 'red')
plt.plot(bot_x, bot_sr1.predict(bot_x), color = 'blue')
plt.title('Turtlebot3 for Green carpet')
plt.xlabel('Speed in m/s')
plt.ylabel('Standard Deviation')
plt.show()
```

```python
plt.scatter(bot_x, bot_y2, color = 'red')
plt.plot(bot_x, bot_sr2.predict(bot_x), color = 'blue')
plt.title('Turtlebot3 for Floor')
plt.xlabel('Speed in m/s')
plt.ylabel('Standard Deviation')
plt.show()
```

## Polynominal Regression


```python
from sklearn.preprocessing import PolynomialFeatures
bot_preg1 = PolynomialFeatures(degree = 3)
bot_pr1 = bot_preg1.fit_transform(bot_x)
bot_sreg1 = LinearRegression()
bot_sreg1.fit(bot_pr1, bot_y1)
```

```python
from sklearn.preprocessing import PolynomialFeatures
bot_preg2 = PolynomialFeatures(degree = 7)
bot_pr2 = bot_preg2.fit_transform(bot_x)
bot_sreg2 = LinearRegression()
bot_sreg2.fit(bot_pr2, bot_y2)
```

```python
plt.scatter(bot_x, bot_y1, color = 'red')
plt.plot(bot_x, bot_sreg1.predict(bot_preg1.fit_transform(bot_x)), color = 'blue')
plt.title('Turtlebot3 for Green carpet')
plt.xlabel('Speed in m/s')
plt.ylabel('Standard Deviation')
plt.show()
```

```python
plt.scatter(bot_x, bot_y2, color = 'red')
plt.plot(bot_x, bot_sreg2.predict(bot_preg2.fit_transform(bot_x)), color = 'blue')
plt.title('Turtlebot3 for floor')
plt.xlabel('Speed in m/s')
plt.ylabel('Standard Deviation')
plt.show()
```
